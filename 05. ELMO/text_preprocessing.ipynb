```python
with open('data/test_data_cleaned.txt') as f:
    data = f.readlines()
    
data
```

```python
from tqdm.notebook import tqdm

data_clean = []

for line in tqdm(data):
    line = line.replace(",", " ").replace("\n", "").replace("\t", "")
    sample = [l.lstrip() for l in line.split(".")]
    data_clean.extend(sample)
```

```python
import pandas as pd

df = pd.DataFrame(data_clean)
df
```

```python
df.to_csv('data/test_data_cleaned.txt', index=False, sep='\t', header=None)
```
